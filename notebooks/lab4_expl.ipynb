# Experiment 1

```python
obs_lat, obs_lon = 40, 200
sigma_obs = np.sqrt(10)
R = xr.DataArray([[sigma_obs**2]], coords={"ob": [0], "ob2": [0]})
```

```python
ds = dss[year]

x_true = ds.z500.sel(lat=obs_lat, lon=obs_lon, ensemble=0).item()
x_true
```

```python
y = x_true + norm(0, sigma_obs).rvs(random_state=default_rng(int(1415456754 + year)))
y
```

```python
prior_samples = stack_state(ds.sel(ensemble=slice(1, None)))
prior_samples
```

```python
Ne = len(prior_samples.ensemble)
Ne
```

```python
x_hat = prior_samples.mean("ensemble")
x_hat
```

```python
X = prior_samples - x_hat
X
```

```python
sigma_p_hat_sq = ((Z @ Z.T) / (Ne - 1)).item()
sigma_p_hat_sq
```

```python
H = xr.DataArray(np.zeros((1, len(X.elem))), coords={"ob": [0], "elem": X.elem})
H.loc[dict(lat=obs_lat, lon=obs_lon, field="z500")] = 1
H.sum()
```

```python
Z = H @ X
Z
```

```python
C =Z.rename({"ensemble": "ensemble1"}).T @ Z.rename({"ensemble": "ensemble2"}) / ((sigma_obs**2 + sigma_p_hat_sq) * (Ne - 1)**2)
C
```

```python
alpha = 1 / (1 + np.sqrt(sigma_obs**2 / (sigma_obs**2 + sigma_p_hat_sq)))
T = np.identity(len(X.ensemble)) - alpha * C
T
```

```python
Xa = (X @ T.rename({"ensemble1": "ensemble"})).rename({"ensemble2": "ensemble"})
Xa
```

```python
D_hat = sigma_p_hat_sq + R
D_hat
```

```python
x_hat_a = (x_hat + (X @ Z.T @ inverse(D_hat)) /(Ne - 1) @ (y - H @ x_hat)).squeeze()
x_hat_a
```

```python
fig, axs = plt.subplots(2, 2, subplot_kw=dict(projection=ccrs.EqualEarth(central_longitude=obs_lon)))
((ax_z_prior, ax_u_prior), (ax_z_post, ax_u_post)) = axs

plot_field([ax_z_prior, ax_z_post], [unstack_state(x_hat), unstack_state(x_hat_a)], "z500")
plot_field([ax_u_prior, ax_u_post], [unstack_state(x_hat), unstack_state(x_hat_a)], "u500")

for ax in axs.flat:
    ax.coastlines()
    ax.scatter(obs_lon, obs_lat)
```
